```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def creutzConvert(filename):
    f= open(filename)
    content = f.readlines()
    x = []
    y = []
    error = []
    for i in content:
        x.append(i.split(' ')[0])
        y.append(i.split(' ')[1])
        error.append(i.split(' ')[2])
    
    for i in range(len(error)):
        error[i] = error[i].strip('\n')
        x[i] = float(x[i])
        y[i] = float(y[i])
        error[i] = float(error[i])
        f.close()
    return x,y,error
```

```python
x,y,error = creutzConvert("src/string2goMHMC0.txt")
# x1,y1,error1 = creutzConvert("src/string2goMHMC1.txt")
# x2,y2,error2 = creutzConvert("src/string2goMHMC2.txt")
```

```python
fig, ax = plt.subplots(figsize=(8, 10))
ax.errorbar(x,y,error,fmt='.',label="$\chi(1,1$")
# ax.errorbar(x1,y1,error1,fmt='*')
# ax.errorbar(x2,y2,error2,fmt='x')
ax.plot(x,-np.log(np.array(x)/8),label="$\ln(g^2/2)$")
plt.yscale('log')
yticks = [1, 2,3]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=13);
ax.set_xlabel(r"$\beta$",fontsize=13)
ax.set_ylabel("$\chi(R,R)$",fontsize=13)
ax.legend()
plt.show()
```
